<h4 align='center'><b>Data Cleaning</b></h4>

```python
print(data.isnull().sum())
```

```python
# Fill missing Age with median
data['age'] = data['age'].fillna(data['age'].median())
```

```python
# Fill Embarked with mode
data['embarked'] = data['embarked'].fillna(data['embarked'].mode())
```

```python
# Drop Deck (too many missing values)
data.drop(columns=['deck'], inplace=True)
```

```python
# Fill embark_town with mode
data['embark_town'].fillna(data['embark_town'].mode())
```

```python
data
```

#### datatype conversion

```python
data['sex'] = data['sex'].astype('category')
type('sex')
```

```python
data['embarked'] = data['embarked'].astype('category')
type('embarked')
```

<h4 align = 'center'><b>Summary Statistics and Group Analysis</b></h4>

```python
# Survival rate by gender
print(data.groupby('sex')['survived'].mean())
```

```python
# Survival rate by class
print(data.groupby('pclass')['survived'].mean())
```

<h4 align='center'><b>Visualization</b></h4>

#### survival gender wise

```python
sns.barplot(x='sex',y='survived',data=data,color='g')
plt.title('Survival Rate by Gender')
plt.show()
```

#### **survival class wise**

```python
sns.barplot(x='pclass',y='survived',data=data,color='y')
plt.title('Survival Rate by class')
plt.show()
```

#### **Age distribution by survival**

```python
sns.histplot(data=data,x='age',hue='survived',multiple='stack',kde=True,palette='Set1')
plt.title('Age distribution by survival')
plt.show()
```

#### **Feature correlation**

```python
sns.heatmap(data.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('Feature Correlation')
plt.show()
```

### **Insights:**

##### **Gender and Survival**:
Females had a significantly higher survival rate than males.\
Most women and children were prioritized during evacuation.
##### **Passenger Class**:
Passengers in 1st class had the highest survival rate.\
3rd class passengers had the lowest chance of survival.
##### **Age Distribution**:
Children under 10 had relatively high survival rates.\
Elderly passengers had lower survival chances.
##### **Fare**:
Survivors tended to have paid higher fares on average.\
Indicates wealth or class was a key survival factor.
##### **Embarkation Port**:
Most passengers boarded from Southampton (S).\
Survival rates varied slightly based on embarkation location.


### **Conclusion**:
Survival on the Titanic was influenced heavily by gender, class, and age. Females, especially in first class, were far more likely to survive than males or passengers in lower classes.
